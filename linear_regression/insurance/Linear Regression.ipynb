```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['font.size'] =14
plt.rcParams['font.weight']= 'bold'
plt.style.use('seaborn-whitegrid')
```

```python
data = pd.read_csv('insurance.csv', low_memory = False)
data.head()
```

```python
data.describe(include = 'all')
```

```python
plt.figure(figsize=(12,4))
sns.heatmap(data.isnull(),yticklabels= False, cbar = False, cmap = 'Greens_r')
plt.title('Missing value in the dataset')
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(data.corr(),cmap = 'coolwarm_r', annot = True)
plt.title('correlation between variables')
```

```python
f, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.kdeplot(data['charges'],color = 'r', fill = True, ax = axs[0]).set_title('distrubution of charges')
sns.kdeplot(np.log10(data['charges']),color = 'b', fill = True, ax = axs[1]).set_title('distrubution of log charges')
```

```python
data['log_charges'] = np.log10(data['charges'])
data.head()
```

```python
f, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.kdeplot(data = data, x = 'sex', y = 'charges', color = 'r', fill = True, hue = 'smoker', ax = axs[0]).set_title('distrubution of male and female')
#sns.kdeplot(np.log10(data['charges']),color = 'b', fill = True, ax = axs[1]).set_title('distrubution of log charges')
```
